```python
#Лабораторная работа 3
#Выполнила: Михалева Полина Вячеславовна, ЦИБ-251
#Задача 17: Реализуйте функцию insertion_sort_optimized(arr) (обычная сортировка вставками). Тест проверит, правильно ли она работает на массиве, где переставлены только два элемента.

def insertion_sort_optimized(arr):
    """
    Алгоритм: сортировка вставками
    Сложность: O(N^2) в худшем случае, O(N) в лучшем
    """
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1

        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1

        arr[j + 1] = key

    return arr


# Тестирование
if __name__ == "__main__":
    a = [1, 2, 3, 5, 4]
    insertion_sort_optimized(a)
    print(f"Тест 1: {a}")

    b = [10, 20, 30, 25, 40, 50]
    insertion_sort_optimized(b)
    print(f"Тест 2: {b}")

    c = [1, 2, 3, 4, 5]
    insertion_sort_optimized(c)
    print(f"Тест 3: {c}")
```
