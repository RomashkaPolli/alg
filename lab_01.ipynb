```python
from typing import Any
class Queue:
    """Базовая реализация очереди (FIFO) на основе списка Python."""

    def __init__(self) -> None:
        self._items: list[Any] = []

    def enqueue(self, item: Any) -> None:
        """Добавить элемент в конец очереди. O(1)"""
        self._items.append(item)

    def dequeue(self) -> Any:
        """Удалить и вернуть первый элемент. O(n) для списка Python."""
        if self.is_empty():
            raise IndexError("Невозможно удалить элемент из пустой очереди")
        return self._items.pop(0)

    def peek(self) -> Any:
        if self.is_empty():
            raise IndexError("Невозможно просмотреть элемент пустой очереди")
        return self._items[0]

    def is_empty(self) -> bool:
        return len(self._items) == 0

    def size(self) -> int:
        return len(self._items)
```

```python
class MovingAverage:
    """Скользящее среднее для потока чисел с фиксированным окном."""

    def __init__(self, window_size: int) -> None:
        if window_size <= 0:
            raise ValueError("Размер окна должен быть положительным числом (>0)")
        self.window_size: int = window_size
        self.queue: Queue = Queue()
        self.current_sum: float = 0.0

    def next(self, val: float) -> float:
        """O(1)"""
        if self.queue.size() == self.window_size:
            removed: float = self.queue.dequeue()
            self.current_sum -= removed

        self.queue.enqueue(val)
        self.current_sum += val

        return self.current_sum / self.queue.size()

def main() -> None:
    window = 3
    ma = MovingAverage(window)
    stream = [1, 10, 3, 5, 2, 8, 4]

    print(f"Скользящее среднее (окно = {window}):")
    print("-" * 45)
    for num in stream:
        avg = ma.next(num)
        print(f"Вход: {num:2} | Окно: {str(ma.queue._items):<12} | Среднее: {avg:.2f}")
    print("-" * 45)

    print("\nПроверка граничных условий:")
    try:
        bad_ma = MovingAverage(0)
    except ValueError as e:
        print(f"Корректно перехвачено: {e}")

    try:
        q = Queue()
        q.dequeue()
    except IndexError as e:
        print(f"Корректно перехвачено: {e}")

if __name__ == "__main__":
    main()
```
